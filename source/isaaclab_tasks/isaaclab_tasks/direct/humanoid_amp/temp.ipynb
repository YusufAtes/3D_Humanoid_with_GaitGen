```python
import numpy as np

data = np.load("plane_0.25_3.npy")
# Reshape the data to 32x1024
arr_32_1024 = data.reshape(32, 1024)

# For each row in arr_32_1024, create 32 copies (making all rows identical in each 32-row block)
# This makes a (32*32, 1024) = (1024, 1024) array
arr_1024_1024 = np.repeat(arr_32_1024, 32, axis=0)

# Save the resulting array as "plane_0.npy"
np.save("plane_3.npy", arr_1024_1024)

# arr_1024_1024 now has the desired shape with each 32-row block being copies of each original row
print(arr_1024_1024.shape)

```
